```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 3*x - 2
```

```python
# Plot the function

x = np.linspace(-1, 1, 200)
y = f(x)
plt.plot(x, y)
plt.show()
```

```python
def g(x):
    return 3*x**2 - 2 
```

```python
# Plot the function

xp = np.linspace(-1, 1, 200)
yp = g(xp)
plt.plot(xp, yp)
plt.show()
```

```python
a, g_a = -1, g(-1)
```

```python
plt.plot(xp, yp)
plt.scatter(a, g_a)
plt.show()
```

```python
def derivative(f, p):
    """
    f: function that we will differentiate
    p: point at which we will evaluate the derivative
    """
    h = 0.00001
    return (f(p + h) - f(p)) / h
```

```python
dxp = np.linspace(-1, 1, 200)
dyp = derivative(g, xp)
plt.plot(dxp, dyp)
plt.grid(True)
plt.show()
```

```python
def gradient_descent(initial_x,learning_rate, epochs):
    
    """This function finds the minimum value of f via 
    the gradient descent algorithm."""

    i = 0
    x = initial_x
    
    while i < epochs:
        dg_x = derivative(g, x)
        x = x - learning_rate * dg_x
        i += 1

    return x, g(x)
```

```python
x_point, y_point = gradient_descent(a,0.001,100)
```

```python
plt.plot(xp,yp, label = 'f(x)')
plt.scatter(x_point, y_point, color='red', zorder=5, label=f'Point at x={x_point}')
plt.grid(True)

plt.show()
```

```python
epochs = 200

while epochs < 1001:
    x_point, y_point = gradient_descent(a,0.001,epochs)
    plt.plot(xp,yp, label = 'f(x)')
    plt.scatter(x_point, y_point, color='red', zorder=5, label=f'Point at x={x_point}')
    plt.grid(True)
    plt.show()

    epochs += 100
```
