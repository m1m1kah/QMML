## Simple Implementation of a Gradient Descent Algorithm ##

In this notebook, we will implement a basic version of the gradient descent algorithm to minimize a simple mathematical function. The process will involve the following steps:

1. **Define the function**: We will define a function \( f(x) \) that we aim to minimize.
2. **Visualize the function**: We will plot the function to understand its behavior.
3. **Implement gradient descent**: We will compute the gradient (derivative) of the function and iteratively update our parameter \( x \) to minimize \( f(x) \).
4. **Plot the results**: Finally, we will plot the path taken by gradient descent to find the minimum.

This will provide a simple and intuitive way to understand how gradient descent works in practice.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define our loss function

def f(x):
    return (x-2)**2 + 1 / 2 * np.sin(x)

# Define initial point

p0 = 0

# Return f(x) evaluated at p0

print(f"f(x) evaluated at x = 0 is {f(p0)}.")
                                
```

```python
# Define the derivative of the loss function

def f_d1(x):
    return 2*(x-2) + 1 / 2 * np.cos(x)
```

```python
def gradient_descent(initial_x,learning_rate, epochs):
    
    """This function finds the minimum value of f via 
    the gradient descent algorithm."""

    i = 0
    x = initial_x
    
    while i < epochs:
        df_x = f_d1(x)
        x = x - learning_rate * df_x
        if i % 500 == 0:
            print(f"Iteration number: {i} x = {x}, f(x) = {f(x)}")
        i += 1

    return x, f(x)
```

```python
# Plot results on the graph

x_point, y_point = gradient_descent(p0,0.001,4001)
```
